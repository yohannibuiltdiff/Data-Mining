```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heights.csv")
```

```python
df
```

```python
df.head()
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

```python
df['earn'].fillna(df['earn'].mean(), inplace=True)
df['height'].fillna(df['height'].mean(), inplace=True)
df['sex'].fillna(df['sex'].mode()[0], inplace=True)
df['ed'].fillna(df['ed'].mode()[0], inplace=True)
df['age'].fillna(df['age'].mean(), inplace=True)
df['race'].fillna(df['race'].mode()[0], inplace=True)
```

```python
df['age'] = df['age'].astype(int)
df['race'] = df['race'].astype('category')
df['earn'] = df['earn'].astype(int)
df['height'] = df['height'].astype(int)
df['sex'] = df['sex'].astype('category')
df['ed'] = df['ed'].astype(int)
```

```python
print("\nUnique values in 'earn':", df['earn'].unique())
print("Unique values in 'height':", df['height'].unique())
print("Unique values in 'sex':", df['sex'].unique())
print("\nUnique values in 'ed':", df['ed'].unique())
print("Unique values in 'age':", df['age'].unique())
print("Unique values in 'race':", df['race'].unique())
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
print("\nCleaned DataFrame:\n", df)
```
