```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diamonds.csv")
```

```python
df
```

```python
df.head()
```

```python
print("Missing values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
print("\nSummary statistics:\n", df.describe())
```

```python
duplicate_rows = df[df.duplicated()]
print("\nDuplicate rows:\n", duplicate_rows)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['carat'], bins=10, kde=True)
plt.title('Distribution of Carat')
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df['price'], bins=10, kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='carat', y='price')
plt.title('Scatter Plot of Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df['cut'] = df['cut'].astype('category')
df['color'] = df['color'].astype('category')
df['clarity'] = df['clarity'].astype('category')
```

```python
print("\nUnique values in 'cut':", df['cut'].unique())
print("Unique values in 'color':", df['color'].unique())
print("Unique values in 'clarity':", df['clarity'].unique())
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
print("\nCleaned DataFrame:\n", df)
```
